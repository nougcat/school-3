```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
litery = pd.read_csv('litery.csv')
litery = litery.head(35)
default = pd.read_csv('litery_default.csv')
```

```python
while litery.iloc[3,0] <= default.iloc[3,0]:
   default = default.div(1.1)
```

```python
plt.plot(litery, color='green')
plt.plot(default, color='red')
plt.ylabel('częstotliwość występowania')
plt.xlabel('litery alfabetu')
plt.show()
```

Wykresy te najprawdopodobniej nie pasują do siebie ponieważ miałem nie reprezentatywne dane - nie pokazują one najcześciej występujących liter w języku polskim, a te które są najczęściej w słowniku języka polskiego. A to oznacza że długie, stare, nie używane słowa są de facto przedstawione na tym wykresie, a nie take słowa jakie najczęsciej używamy jak 'ok' czy 'siema'
